```python
# Import packages and set plot parameters
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
from scipy.signal import find_peaks
plt.rcParams['figure.dpi'] = 150
```

```python
# Define gaussian filter subtraction and smoothing function
def gauss_sub(i):
    g = gaussian_filter1d(i,sigma=50)
    isub = i-g
    y = gaussian_filter1d(isub,sigma=2)
    return y

# Define function to get x and y position of peaks
def peaks(r,y,h):
    p = find_peaks(y,height=h,distance=20)
    ph = p[1]['peak_heights']
    pp = r[p[0]]
    return pp,ph
```

```python
# Import data and define variables
b3 = np.genfromtxt('Data/B0959.3c_MapAvr_26-03-04.txt',delimiter='\t')
b3r = b3[:-75,0]
b3i = b3[:-75,1]
b3y = gauss_sub(b3i)
p3p,p3h = peaks(b3r,b3y,75)

b4 = np.genfromtxt('Data/B0959.4c_MapAvr_26-03-04.txt',delimiter='\t')
b4r = b4[:-75,0]
b4i = b4[:-75,1]
b4y = gauss_sub(b4i)
p4p,p4h = peaks(b4r,b4y,50)

b5 = np.genfromtxt('Data/B0959.5a_MapAvr_26-03-06.txt',delimiter='\t')
b5r = b5[:-75,0]
b5i = b5[:-75,1]
b5y = gauss_sub(b5i)
p5p,p5h = peaks(b5r,b5y,100)

b8 = np.genfromtxt('Data/B0959.8c_MapAvr_26-03-06.txt',delimiter='\t')
b8r = b8[:-75,0]
b8i = b8[:-75,1]
b8y = gauss_sub(b8i)
p8p,p8h = peaks(b8r,b8y,50)

b15 = np.genfromtxt('Data/B0959.15a_MapAvr_26-03-18.txt',delimiter='\t')
b15r = b15[:-75,0]
b15i = b15[:-75,1]
b15y = gauss_sub(b15i)
p15p,p15h = peaks(b15r,b15y,150)

b16 = np.genfromtxt('Data/B0959.16b_MapAvr_26-03-18.txt',delimiter='\t')
b16r = b16[:-75,0]
b16i = b16[:-75,1]
b16y = gauss_sub(b16i)
p16p,p16h = peaks(b16r,b16y,27)

k = np.genfromtxt('Data/M08_Kaolin_MapAvr_25-08-14.txt',delimiter='\t')
kr = k[:-75,0]
ki = k[:-75,1]
ky = gauss_sub(ki)
pkp,pkh = peaks(kr,ky,85)
```

```python
plt.subplot(711)
frame1 = plt.gca()
plt.plot(b3r,b3y,c='Black',linewidth=1)
plt.plot(p3p,p3h,"x")
plt.xlim(180,1200)
plt.ylim(-70,170)
frame1.axes.get_yaxis().set_ticks([])
frame1.axes.get_xaxis().set_ticks([])

plt.subplot(712)
frame2 = plt.gca()
plt.plot(b4r,b4y,c='Black',linewidth=1)
plt.plot(p4p,p4h,"x")
plt.xlim(180,1200)
plt.ylim(-50,180)
frame2.axes.get_yaxis().set_ticks([])
frame2.axes.get_xaxis().set_ticks([])

plt.subplot(713)
frame3 = plt.gca()
plt.plot(b5r,b5y,c='Black',linewidth=1)
plt.plot(p5p,p5h,"x")
plt.ylim(-100,600)
plt.xlim(180,1200)
frame3.axes.get_yaxis().set_ticks([])
frame3.axes.get_xaxis().set_ticks([])

plt.subplot(714)
frame4 = plt.gca()
plt.plot(b8r,b8y,c='Black',linewidth=1)
plt.plot(p8p,p8h,"x")
plt.xlim(180,1200)
plt.ylim(-75,180)
frame4.axes.get_yaxis().set_ticks([])
frame4.axes.get_xaxis().set_ticks([])
plt.ylabel('Intensity (Gaussian Filter $\\sigma$=2,\nGaussian Filter $\\sigma$=50 Subtracted)')

plt.subplot(715)
frame5 = plt.gca()
plt.plot(b15r,b15y,c='Black',linewidth=1)
plt.plot(p15p,p15h,"x")
plt.xlim(180,1200)
plt.ylim(-180,600)
frame5.axes.get_yaxis().set_ticks([])
frame5.axes.get_xaxis().set_ticks([])

plt.subplot(716)
frame6 = plt.gca()
plt.plot(b16r,b16y,c='Black',linewidth=1)
plt.plot(p16p,p16h,"x")
plt.xlim(180,1200)
plt.ylim(-30,70)
frame6.axes.get_yaxis().set_ticks([])
frame6.axes.get_xaxis().set_ticks([])

plt.subplot(717)
frame7 = plt.gca()
plt.plot(kr,ky,linewidth=1)
plt.plot(pkp,pkh,"x")
for i,j in zip(pkp,pkh):
    plt.annotate(np.round(i,decimals=1),xy=(i,j),xytext=(5, 5),textcoords='offset points')
plt.xlim(180,1200)
plt.ylim(-100,400)
frame7.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
```
