```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
plt.rcParams['figure.dpi'] = 150
```

```python
B5a = np.genfromtxt('B0959.5a_260306_MAP_20x20_Average.txt',delimiter='\t')
r = B5a[:-50,0]
i = B5a[:-50,1]
```

```python
g = gaussian_filter1d(i,sigma=50)
isub = i-g
y = gaussian_filter1d(isub,sigma=2)
```

```python
ss = 0.5
cc = 'Black'
plt.scatter(r,i,c=cc,s=ss) 
plt.plot(r,i,c=cc,linewidth=ss) 
```

```python
plt.scatter(r,g,c=cc,s=ss) 
plt.plot(r,g,c=cc,linewidth=ss)
```

```python
plt.scatter(r,isub,c=cc,s=ss) 
plt.plot(r,isub,c=cc,linewidth=ss)
plt.xlim(np.min(r),np.max(r))
```

```python
plt.plot(r,y,c=cc,linewidth=1)
plt.xlim(np.min(r),np.max(r))
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Gaussian Filter $\\sigma$=2 Intensity (Gaussian Filter $\\sigma$=50 Subtracted)')
```

```python
di = np.gradient(i)
yy = gaussian_filter1d(-np.gradient(di),sigma=2)
```

```python
plt.plot(r,yy,c='Black',linewidth=1) 
plt.xlim(np.min(r),np.max(r))
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity Gradient (2nd Derivative, Gaussian Filtered $\\sigma$=2')
```
