```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pywt
from scipy.ndimage import gaussian_filter
from scipy.ndimage import gaussian_filter1d
```

```python
def split_2d_wavelet_components(z, wavelet="db2", level=3, keep_coarse_levels=1):
    """
    z : 2D ndarray
    wavelet : 'db2', 'haar', 'sym4', 'coif2'
    level : Total decomposition level.

    z_low : 2D ndarray reconstruction using only coarse scales.
    z_high : 2D ndarray residual fine-scale component.
    coeffs : Full wavelet coefficient structure.
    """
    coeffs = pywt.wavedec2(z, wavelet=wavelet, level=level)

    # coeffs structure:
    # coeffs[0] = approximation at coarsest level
    # coeffs[1:] = detail coeffs from coarse to fine:[(cH_n, cV_n, cD_n), ..., (cH_1, cV_1, cD_1)]

    # Build a coefficient set for low-pass reconstruction
    coeffs_low = []
    # Always keep the coarsest approximation
    coeffs_low.append(coeffs[0])
    n_detail_levels = len(coeffs) - 1

    for i, detail in enumerate(coeffs[1:], start=1):
        # i = 1 is coarsest detail level, larger spatial scale
        # i = n_detail_levels is finest detail level
        if i <= keep_coarse_levels:
            coeffs_low.append(detail)
        else:
            coeffs_low.append(tuple(np.zeros_like(d) for d in detail))

    z_low = pywt.waverec2(coeffs_low, wavelet=wavelet)
    # Match original shape if reconstruction is slightly padded
    z_low = z_low[:z.shape[0], :z.shape[1]]
    z_high = z - z_low
    return z_low, z_high, coeffs
```

```python
# Make synthetic dataset
nx, ny = 256, 256
x = np.arange(nx)
y = np.arange(ny)
X, Y = np.meshgrid(x, y)

z = (
    2.0 * np.sin(2 * np.pi * X / 90.0) +
    1.5 * np.sin(2 * np.pi * Y / 70.0) +
    0.7 * np.sin(2 * np.pi * X / 12.0) * np.sin(2 * np.pi * Y / 10.0)
)
```

```python
# analysis
z_low, z_high, coeffs = split_2d_wavelet_components(z,wavelet="db2",level=4,keep_coarse_levels=2)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4.5), constrained_layout=True)

im0 = axes[0].imshow(z, origin="lower", cmap="terrain")
axes[0].set_title("Original field")
plt.colorbar(im0, ax=axes[0], shrink=0.8)

im1 = axes[1].imshow(z_low, origin="lower", cmap="terrain")
axes[1].set_title("Coarse-scale component")
plt.colorbar(im1, ax=axes[1], shrink=0.8)

im2 = axes[2].imshow(z_high, origin="lower", cmap="terrain")
axes[2].set_title("Fine-scale component")
plt.colorbar(im2, ax=axes[2], shrink=0.8)

for ax in axes:
    ax.set_xlabel("x")
    ax.set_ylabel("y")

plt.show()
```

```python
t2019a = np.genfromtxt('T2019.06.10a_260319_MAP_Average.txt',delimiter='\t')
r = t2019a[:,0]
i = t2019a[:,1]
```

```python
ss = 0.5
cc = 'Black'
plt.scatter(r,i,c=cc,s=ss) 
plt.plot(r,i,c=cc,linewidth=ss) 
plt.xlim(50,1250)
plt.ylim(750,np.max(i)+(np.mean(i)/10))
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity')
```

```python
Gi = gaussian_filter1d(t2019a[:,1], sigma=6)
```

```python
plt.scatter(r,Gi,c=cc,s=ss) 
plt.plot(r,Gi,c=cc,linewidth=ss) 
plt.xlim(50,1250)
plt.ylim(750,1500)
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity (Gaussian Filtered)')
```

```python
DGi = np.gradient(Gi)
```

```python
plt.plot(r,DGi,c=cc,linewidth=1) 
plt.xlim(50,1250)
plt.ylim(-30,30)
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity Gradient (Gaussian Filtered)')
```
