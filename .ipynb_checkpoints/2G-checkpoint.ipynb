```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
# plt.rcParams['figure.dpi'] = 600
```

```python
h = np.genfromtxt('RRUFF_Hematite_R060190_Eng.txt',delimiter=', ')
hr = h[:-75,0]
hi = h[:-75,1]
hdi = np.gradient(hi)
hyy = gaussian_filter1d(-np.gradient(hdi),sigma=2)

b5a = np.genfromtxt('B0959.5a_MapAvr_26-03-06.txt',delimiter='\t')
b5r = b5a[:-75,0]
b5i = b5a[:-75,1]
b5di = np.gradient(b5i)
b5yy = gaussian_filter1d(-np.gradient(b5di),sigma=2)
```

```python
frame1 = plt.gca()
plt.plot(hr,hyy,c='Black',linewidth=1) 
plt.xlim(180,1200)
frame1.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Inverse 2nd Gradient of Intensity \n(Gaussian Filtered $\\sigma$=2)')
plt.title('RRUFF Hematite 2nd Gradient of Raman Spectrum')
# plt.savefig('Hematite_2G.png')
```

```python
frame2 = plt.gca()
plt.plot(b5r,b5yy,c='Black',linewidth=1) 
plt.xlim(180,1200)
frame2.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Inverse 2nd Gradient of Intensity \n(Gaussian Filtered $\\sigma$=2)')
plt.title('B959.5a 2nd Gradient of Raman Spectrum')
# plt.savefig('B0959.5a_2G.png')
```
