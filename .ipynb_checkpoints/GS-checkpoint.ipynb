```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
# plt.rcParams['figure.dpi'] = 600
```

```python
k = np.genfromtxt('RRUFF_Hematite_R060190_Eng.txt',delimiter=', ')
kr = k[:-75,0]
ki = k[:-75,1]
kg = gaussian_filter1d(ki,sigma=50)
kisub = ki-kg
ky = gaussian_filter1d(kisub,sigma=2)

b3 = np.genfromtxt('B0959.3c_MapAvr_26-03-04.txt',delimiter='\t')
b3r = b3[:-75,0]
b3i = b3[:-75,1]
b3g = gaussian_filter1d(b3i,sigma=50)
b3isub = b3i-b3g
b3y = gaussian_filter1d(b3isub,sigma=2)

b4 = np.genfromtxt('B0959.4c_MapAvr_26-03-04.txt',delimiter='\t')
b4r = b4[:-75,0]
b4i = b4[:-75,1]
b4g = gaussian_filter1d(b4i,sigma=50)
b4isub = b4i-b4g
b4y = gaussian_filter1d(b4isub,sigma=2)

b5 = np.genfromtxt('B0959.5a_MapAvr_26-03-06.txt',delimiter='\t')
b5r = b5[:-75,0]
b5i = b5[:-75,1]
b5g = gaussian_filter1d(b5i,sigma=50)
b5isub = b5i-b5g
b5y = gaussian_filter1d(b5isub,sigma=2)

b8 = np.genfromtxt('B0959.8c_MapAvr_26-03-06.txt',delimiter='\t')
b8r = b8[:-75,0]
b8i = b8[:-75,1]
b8g = gaussian_filter1d(b8i,sigma=50)
b8isub = b8i-b8g
b8y = gaussian_filter1d(b8isub,sigma=2)

b12 = np.genfromtxt('B0959.12a_MapAvr_26-03-13.txt',delimiter='\t')
b12r = b12[:-75,0]
b12i = b12[:-75,1]
b12g = gaussian_filter1d(b12i,sigma=50)
b12isub = b12i-b12g
b12y = gaussian_filter1d(b12isub,sigma=2)

b14 = np.genfromtxt('B0959.14a_MapAvr_26-03-18.txt',delimiter='\t')
b14r = b12[:-75,0]
b14i = b12[:-75,1]
b14g = gaussian_filter1d(b14i,sigma=50)
b14isub = b14i-b14g
b14y = gaussian_filter1d(b14isub,sigma=2)

b15 = np.genfromtxt('B0959.15a_MapAvr_26-03-18.txt',delimiter='\t')
b15r = b15[:-75,0]
b15i = b15[:-75,1]
b15g = gaussian_filter1d(b15i,sigma=50)
b15isub = b15i-b15g
b15y = gaussian_filter1d(b15isub,sigma=2)

b16 = np.genfromtxt('B0959.16b_MapAvr_26-03-18.txt',delimiter='\t')
b16r = b16[:-75,0]
b16i = b16[:-75,1]
b16g = gaussian_filter1d(b16i,sigma=50)
b16isub = b16i-b16g
b16y = gaussian_filter1d(b16isub,sigma=2)
```

```python
frame1 = plt.gca()
plt.plot(kr,ky,c='Black',linewidth=1)
plt.xlim(180,1200)
frame1.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity (Gaussian Filter $\\sigma$=2, \nGaussian Filter $\\sigma$=50 Subtracted)')
plt.title('M08 Kaolinite Gaussian Subtracted Raman Spectrum 25-08-14')
# plt.savefig('Kaolinite_GS.png')
```
