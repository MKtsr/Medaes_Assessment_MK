```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
plt.rcParams['figure.dpi'] = 600
```

```python
B5a = np.genfromtxt('B0959.5a_MapAvr_26-03-06.txt',delimiter='\t')
r = B5a[:-75,0]
i = B5a[:-75,1]
g = gaussian_filter1d(i,sigma=50)
isub = i-g
y = gaussian_filter1d(isub,sigma=2)
di = np.gradient(i)
yy = gaussian_filter1d(-np.gradient(di),sigma=2)
```

```python
cc = 'Black'
plt.plot(r,i,c=cc,linewidth=1) 
plt.xlim(180,1200)
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity')
plt.savefig('B0959.5a_Raw.png')
```

```python
frame1 = plt.gca()
plt.plot(r,y,c=cc,linewidth=1)
plt.xlim(180,1200)
frame1.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity (Gaussian Filter $\\sigma$=2, \nGaussian Filter $\\sigma$=50 Subtracted)')
plt.savefig('B0959.5a_GS.png')
```

```python
frame2 = plt.gca()
plt.plot(r,yy,c='Black',linewidth=1) 
plt.xlim(180,1200)
frame2.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Inverse 2nd Gradient of Intensity \n(Gaussian Filtered $\\sigma$=2)')
plt.savefig('B0959.5a_2G.png')
```
