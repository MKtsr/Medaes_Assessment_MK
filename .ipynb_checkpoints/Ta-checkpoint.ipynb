```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
plt.rcParams['figure.dpi'] = 150
```

```python
t2019a = np.genfromtxt('T2019.06.10a_260319_MAP_Average.txt',delimiter='\t')
r = t2019a[:-50,0]
i = t2019a[:-50,1]
g = gaussian_filter1d(i,sigma=50)
isub = i-g
y = gaussian_filter1d(isub,sigma=2)
```

```python
plt.savefig('SAMPLEID_G_S_Raman_Spectrum_DATE.png')
plt.plot(r,y,c='Black',linewidth=1)
plt.xlim(np.min(r),np.max(r))
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Gaussian Filter $\\sigma$=2 Intensity (Gaussian Filter $\\sigma$=50 Subtracted)')
```

```python
di = np.gradient(i)
yy = gaussian_filter1d(-np.gradient(di),sigma=2)
```

```python
plt.plot(r,yy,c='Black',linewidth=1) 
plt.xlim(np.min(r),np.max(r))
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity Gradient (2nd Derivative, Gaussian Filtered $\\sigma$=2')
```
