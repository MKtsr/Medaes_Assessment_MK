```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
plt.rcParams['figure.dpi'] = 600
```

```python
a = np.genfromtxt('M08_Kaolin_MapAvr_25-08-14.txt',delimiter='\t')
r = a[:-75,0]
i = a[:-75,1]
g = gaussian_filter1d(i,sigma=50)
isub = i-g
y = gaussian_filter1d(isub,sigma=2)
di = np.gradient(i)
yy = gaussian_filter1d(-np.gradient(di),sigma=2)
```

```python
frame1 = plt.gca()
plt.plot(r,y,c='Black',linewidth=1)
plt.xlim(180,1200)
frame1.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity (Gaussian Filter $\\sigma$=2, \nGaussian Filter $\\sigma$=50 Subtracted)')
plt.title('M08 Kaolinite Gaussian Subtracted Raman Spectrum 25-08-14')
plt.savefig('Kaolinite_GS.png')
```
