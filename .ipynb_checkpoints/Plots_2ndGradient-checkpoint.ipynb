```python
# Import packages and set plot parameters
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
# plt.rcParams['figure.dpi'] = 600
```

```python
# Define second gradient function
def second_gradient(i):
    di = np.gradient(i)
    y = gaussian_filter1d(-np.gradient(di),sigma=2)
    return y
```

```python
# Import data and define variables
h = np.genfromtxt('RRUFF_Hematite_R060190_Eng.txt',delimiter=', ')
hr = h[:-75,0]
hi = h[:-75,1]
hy = second_gradient(hi)

b3 = np.genfromtxt('B0959.3c_MapAvr_26-03-04.txt',delimiter='\t')
b3r = b3[:-75,0]
b3i = b3[:-75,1]
b3y = second_gradient(b3i)

b4 = np.genfromtxt('B0959.4c_MapAvr_26-03-04.txt',delimiter='\t')
b4r = b4[:-75,0]
b4i = b4[:-75,1]
b4y = second_gradient(b4i)

b5 = np.genfromtxt('B0959.5a_MapAvr_26-03-06.txt',delimiter='\t')
b5r = b5[:-75,0]
b5i = b5[:-75,1]
b5y = second_gradient(b5i)

b8 = np.genfromtxt('B0959.8c_MapAvr_26-03-06.txt',delimiter='\t')
b8r = b8[:-75,0]
b8i = b8[:-75,1]
b8y = second_gradient(b8i)

b12 = np.genfromtxt('B0959.12a_MapAvr_26-03-13.txt',delimiter='\t')
b12r = b12[:-75,0]
b12i = b12[:-75,1]
b12y = second_gradient(b12i)

b14 = np.genfromtxt('B0959.14a_MapAvr_26-03-18.txt',delimiter='\t')
b14r = b12[:-75,0]
b14i = b12[:-75,1]
b14y = second_gradient(b14i)

b15 = np.genfromtxt('B0959.15a_MapAvr_26-03-18.txt',delimiter='\t')
b15r = b15[:-75,0]
b15i = b15[:-75,1]
b15y = second_gradient(b15i)

b16 = np.genfromtxt('B0959.16b_MapAvr_26-03-18.txt',delimiter='\t')
b16r = b16[:-75,0]
b16i = b16[:-75,1]
b16y = second_gradient(b16i)
```

```python
frame1 = plt.gca()
plt.plot(hr,hy,linewidth=1) 
plt.xlim(180,1200)
frame1.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Inverse 2nd Gradient of Intensity \n(Gaussian Filtered $\\sigma$=2)')
plt.title('RRUFF Hematite 2nd Gradient of Raman Spectrum')
# plt.savefig('Hematite_2G.png')
```
