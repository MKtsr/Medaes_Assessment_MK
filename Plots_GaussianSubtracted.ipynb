```python
# Import packages and set plot parameters
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter1d
# plt.rcParams['figure.dpi'] = 600
```

```python
# Define gaussian filter subtraction function
def gauss_sub(i):
    g = gaussian_filter1d(i,sigma=50)
    isub = i-g
    y = gaussian_filter1d(isub,sigma=2)
    return y
```

```python
# Import data and define variables
k = np.genfromtxt('Data/M08_Kaolin_MapAvr_25-08-14.txt',delimiter='\t')
kr = k[:-75,0]
ki = k[:-75,1]
ky = gauss_sub(ki)

b3 = np.genfromtxt('Data/B0959.3c_MapAvr_26-03-04.txt',delimiter='\t')
b3r = b3[:-75,0]
b3i = b3[:-75,1]
b3y = gauss_sub(b3i)

b4 = np.genfromtxt('Data/B0959.4c_MapAvr_26-03-04.txt',delimiter='\t')
b4r = b4[:-75,0]
b4i = b4[:-75,1]
b4y = gauss_sub(b4i)

b5 = np.genfromtxt('Data/B0959.5a_MapAvr_26-03-06.txt',delimiter='\t')
b5r = b5[:-75,0]
b5i = b5[:-75,1]
b5y = gauss_sub(b5i)

b8 = np.genfromtxt('Data/B0959.8c_MapAvr_26-03-06.txt',delimiter='\t')
b8r = b8[:-75,0]
b8i = b8[:-75,1]
b8y = gauss_sub(b8i)

b12 = np.genfromtxt('Data/B0959.12a_MapAvr_26-03-13.txt',delimiter='\t')
b12r = b12[:-75,0]
b12i = b12[:-75,1]
b12y = gauss_sub(b12i)

b14 = np.genfromtxt('Data/B0959.14a_MapAvr_26-03-18.txt',delimiter='\t')
b14r = b12[:-75,0]
b14i = b12[:-75,1]
b14y = gauss_sub(b14i)

b15 = np.genfromtxt('Data/B0959.15a_MapAvr_26-03-18.txt',delimiter='\t')
b15r = b15[:-75,0]
b15i = b15[:-75,1]
b15y = gauss_sub(b15i)

b16 = np.genfromtxt('Data/B0959.16b_MapAvr_26-03-18.txt',delimiter='\t')
b16r = b16[:-75,0]
b16i = b16[:-75,1]
b16y = gauss_sub(b16i)
```

```python
frame1 = plt.gca()
plt.plot(kr,ky,linewidth=1)
plt.xlim(180,1200)
frame1.axes.get_yaxis().set_ticks([])
plt.xlabel('Raman Shift (cm-1)')
plt.ylabel('Intensity (Gaussian Filter $\\sigma$=2, \nGaussian Filter $\\sigma$=50 Subtracted)')
plt.title('M08 Kaolinite Gaussian Subtracted Raman Spectrum 25-08-14')
# plt.savefig('Kaolinite_GS.png')
```
